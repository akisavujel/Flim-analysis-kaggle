```python
plt.figure (figsize = (6,4)) # Length and breath of box
sns.boxplot (x = df["Budget"]) # Type of diagram

plt.title("Budget outliers box plot") # Giving title to plot diagram
plt.show() # Function that shows plot diagram
```


<div style="display:flex; align-items:center; gap:14px; padding:16px 20px; background:#12091f; border:1px solid rgba(245,166,35,0.2); border-left:4px solid #F5A623; border-radius:12px; margin:20px 0; font-family:sans-serif;">
  <h2 style="margin:0; font-size:20px; font-weight:700; color:#F5A623; letter-spacing:1px;">Interquartile range method</h2>
</div>

```python
q1 = df["Budget"].quantile(0.25) # Quantile of q1
q3 = df["Budget"].quantile(0.75) # Quantile of q2
IQR = q3 - q1 # Formula
```

```python
lowerlimit = q1 - 1.5 * IQR # Lower outlier will be considered anything under first quantile formula
upperlimit = q3 + 1.5 * IQR # upper outlier will be considered anything over third quantile formula
```

```python
df = df[(df["Budget"] >= lowerlimit) & (df["Budget"] <= upperlimit)] # Condition for new output
```

```python
df # Dataframe after removing the outliers
```

```python
import numpy as np

fig, axes = plt.subplots(3, 2, figsize=(16, 20))
fig.patch.set_facecolor("#1a0a2e")

GOLD  = "#F5A623"
PINK  = "#E91E8C"
BG    = "#12091f"
MUTED = "#9b8ab0"

def style_ax(ax, title):
    ax.set_facecolor(BG)
    ax.set_title(title, color=GOLD, fontsize=11, fontweight="bold", pad=12, fontfamily="monospace")
    ax.tick_params(colors=MUTED, labelsize=8)
    ax.xaxis.label.set_color(MUTED)
    ax.yaxis.label.set_color(MUTED)
    for spine in ax.spines.values():
        spine.set_edgecolor(GOLD)
        spine.set_linewidth(0.5)
        spine.set_alpha(0.3)

# Budget outliers box plot
sns.boxplot(x=df["Budget"], ax=axes[0, 0])
style_ax(axes[0, 0], "Budget outliers box plot")

# budget distribution histplot
sns.histplot(df["Budget"], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].ticklabel_format(style="plain", axis="x")
axes[0, 1].set_xlabel("budget")
axes[0, 1].set_ylabel("distribution")
style_ax(axes[0, 1], "budget distribution histplot")

# biggest budget bar graph
movie_budget = df.groupby("Movie")["Budget"].mean().sort_values(ascending=False).head(10)
axes[1, 0].bar(range(len(movie_budget)), movie_budget.values)
axes[1, 0].set_xticks(range(len(movie_budget)))
axes[1, 0].set_xticklabels(movie_budget.index, rotation=45, ha="right", fontsize=7)
axes[1, 0].set_xlabel("budget")
style_ax(axes[1, 0], "biggest budget bar graph")

# biggest overseas boxoffice by movie bar graph
movie_seas = df.groupby("Movie")["Overseas"].mean().sort_values(ascending=False).head(7)
axes[1, 1].bar(range(len(movie_seas)), movie_seas.values)
axes[1, 1].set_xticks(range(len(movie_seas)))
axes[1, 1].set_xticklabels(movie_seas.index, rotation=45, ha="right", fontsize=7)
axes[1, 1].set_xlabel("movies")
style_ax(axes[1, 1], "biggest overseas boxoffice by movie bar graph")

# Common verdict
common_verdict = df["Verdict"].value_counts().head(10)
axes[2, 0].barh(common_verdict.index, common_verdict.values, edgecolor="red")
axes[2, 0].set_xlabel("Number")
axes[2, 0].set_ylabel("Number of verdict")
style_ax(axes[2, 0], "Common verdict")

# India Net vs India Gross
clean = df[["India Net", "India Gross"]].dropna()
axes[2, 1].scatter(clean["India Net"], clean["India Gross"], alpha=0.5, s=18)
axes[2, 1].ticklabel_format(style="plain")
axes[2, 1].set_xlabel("India Net")
axes[2, 1].set_ylabel("India Gross")
corr = clean["India Net"].corr(clean["India Gross"])
style_ax(axes[2, 1], f"India Net vs India Gross  (r = {corr:.2f})")

plt.tight_layout()
plt.savefig("dashboard.png", dpi=150, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.show()
```

## 📊 Business Insights — Bollywood Box Office

### 1. Budget Distribution
- Most Bollywood movies are made with **low to moderate budgets**
- The distribution is heavily **right skewed** — very few films get big budgets
- The industry is dominated by **small budget productions**

### 2. Biggest Budget Movies
- Only a **handful of movies** receive massive budgets
- The top budgeted films are likely **franchise or star-driven** productions
- There is a **huge gap** between the top 2-3 films and the rest

### 3. Overseas Box Office
- Very **few movies dominate** overseas earnings
- The top 3 films earn **significantly more** than the remaining 4
- Overseas success is **not common** — only blockbusters break through internationally

### 4. Common Verdict
- **Flop is the most common verdict** — most Bollywood movies lose money
- Only a small percentage achieve **Hit or Blockbuster** status
- Bollywood is an **extremely high risk** business

### 5. India Net vs India Gross
- Correlation of **r = 0.97** means they are almost perfectly related
- Both metrics move together — **no need to track both** separately
- Domestic performance is **very predictable** once you know either value

---

## 💡 Key Business Takeaway

> **Most Bollywood movies flop. Overseas success is rare. Only a few big budget star-driven films dominate both domestic and international markets.**
